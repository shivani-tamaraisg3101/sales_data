```python
# =====================================
# ðŸ“Š Sales Data Analysis Project
# Description:
#   - Generates sample sales data
#   - Calculates total sales
#   - Visualizes sales by product & over time
# =====================================

# =====================
# 1ï¸âƒ£ Import Libraries
# =====================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better plot style
sns.set(style="whitegrid")

# =====================
# 2ï¸âƒ£ Create Dummy Sales Data
# =====================
# NOTE: Replace this block with pd.read_csv("your_file.csv") if you have real data
data = {
    'Date': pd.date_range(start='2025-01-01', periods=10, freq='D'),
    'Product': ['A', 'B', 'C', 'A', 'B', 'A', 'C', 'B', 'A', 'C'],
    'Units Sold': [10, 15, 7, 20, 14, 18, 5, 22, 25, 12],
    'Price per Unit': [100, 150, 200, 100, 150, 100, 200, 150, 100, 200]
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# =====================
# 3ï¸âƒ£ Calculate Total Sales
# =====================
df['Total Sales'] = df['Units Sold'] * df['Price per Unit']

# Display first rows of dataset
print("ðŸ“Š Sales Data:")
print(df)

# =====================
# 4ï¸âƒ£ Total Sales by Product
# =====================
sales_by_product = df.groupby('Product')['Total Sales'].sum()

print("\nðŸ’° Total Sales by Product:")
print(sales_by_product)

# =====================
# 5ï¸âƒ£ Visualization: Total Sales by Product
# =====================
plt.figure(figsize=(6,4))
sns.barplot(x=sales_by_product.index, y=sales_by_product.values, palette="viridis")
plt.title("Total Sales by Product", fontsize=14)
plt.ylabel("Total Sales ($)", fontsize=12)
plt.xlabel("Product", fontsize=12)
plt.show()

# =====================
# 6ï¸âƒ£ Visualization: Sales Trend Over Time
# =====================
plt.figure(figsize=(8,4))
sns.lineplot(x='Date', y='Total Sales', data=df, marker='o', color='coral')
plt.title("Sales Trend Over Time", fontsize=14)
plt.ylabel("Total Sales ($)", fontsize=12)
plt.xlabel("Date", fontsize=12)
plt.xticks(rotation=45)
plt.show()

```
